#### **Exerc√≠cio 5**

Com base no dicion√°rio a seguir, encontre e retorne uma lista de todas as chaves que tem o valor `True`

```python
# Exerc√≠cio 5
valor = True
dados_loja = {
    'nome': 'Loja de Livros Fant√°sticos',
    'faturamento': 150000.00,
    'estoque_baixo': True,
    'cidades_atendidas': ['S√£o Paulo', 'Curitiba', 'Cuiab√°', 'Manaus', 'Recife'],
    'meta_atingida': False,
    'ano_fundacao': 2010,
    'loja_online': True
}
def buscar_valor(input_dict, input_valor):
  chaves_com_valor = []
  for chave, valor in input_dict.items():
    if valor == input_valor:
      chaves_com_valor.append(chave)
  return chaves_com_valor


# Exibindo o resultado do exerc√≠cio 5
print(buscar_valor(dados_loja, valor))
```

```python
dicionario_teste = { 'valor1': True, 'valor2': False}
assert buscar_valor(dicionario_teste, True) == ['valor1']
print('Teste passou!')
```
